# Question 1

```python
df1=df.copy()
df1['Order Date']=pd.to_datetime(df1['Order Date'])
df1=df1.set_index(['Order Date'])
df1.drop_duplicates(subset ="Customer ID",keep = "first", inplace = True)
df1
```

```python
df1["2011-01-01"]["Profit"].sum()
```

```python
c_11,c_12,c_13,c_14=0,0,0,0
start_date = datetime.date(2011, 1, 1)
end_date = datetime.date(2014, 12, 31)
delta = datetime.timedelta(days=1)

while start_date <= end_date:
    sr=str(start_date)
    if sr[:4]=="2011":
        c_11+=df1.loc[sr].shape[0]
    elif sr[:4]=="2012":
        c_12+=df1.loc[sr].shape[0]
    elif sr[:4]=="2013":
        c_13+=df1.loc[sr].shape[0]
    elif sr[:4]=="2014":
        c_14+=df1.loc[sr].shape[0]
        
    start_date += delta
    
```

```python
import matplotlib.pyplot as plt

w=['2011','2012','2013','2014']
e=[c_11,c_12,c_13,c_14]

plt.plot(w,e,color='maroon')
plt.bar(w,e)
plt.xlabel("Days")
plt.ylabel("No. of costumers")
plt.title("Customer Acquisition vs Days")
plt.show()
```
